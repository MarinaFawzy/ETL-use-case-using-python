```python


import pandas as pd
orders_products_details = pd.read_csv(r'D:\Datamamagemet\python\source\bikestore_dim\Information Mart\orders_products_details.csv')
store_df=pd.read_csv(r'D:\Datamamagemet\python\source\bikestore_dim\staging_2\store\store.csv')
new_orders_products_details=pd.merge(orders_products_details,store_df,on='store_id',suffixes=('_orders', '_other'))
new_orders_products_details
```

```python
import matplotlib.pyplot as plt

new_orders_products_details['total_sales'] = orders_products_details['quantity'] * orders_products_details['list_price_y']

plt.bar(new_orders_products_details['city'], new_orders_products_details['total_sales'])
plt.title('Total Sales per City')
plt.xlabel('City name')
plt.ylabel('Total Sales ($)')
plt.show()
```
