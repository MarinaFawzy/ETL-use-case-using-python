```python


import pandas as pd
orders_products_details = pd.read_csv(r'D:\Datamamagemet\python\source\bikestore_dim\Information Mart\orders_products_details.csv')
orders_products_details.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

orders_products_details['order_date'] = pd.to_datetime(orders_products_details['order_date'])

yearly_order_count = orders_products_details.groupby(orders_products_details['order_date'].dt.year)['order_id'].nunique()

# Plot the line chart
plt.plot(yearly_order_count.index, yearly_order_count.values, marker='o', color='b')
plt.title('Number of Orders Over Time')
plt.xlabel('Year')
plt.ylabel('Count of Orders')
plt.xticks(yearly_order_count.index)

plt.show()
```
