```python
import pandas as pd
import os
# Read exchange data from CSV file into DataFrame
exchange_df = pd.read_csv(r'D:\Datamamagemet\python\source\bikestore_dim\staging_1\exchange\exchange.csv')

# Read order item data from CSV file into DataFrame
order_item_df = pd.read_csv(r'D:\Datamamagemet\python\source\bikestore_dim\staging_1\order_item\order_item.csv')

```

```python
# Find the maximum timestamp value in the exchange DataFrame
max_timestamp = exchange_df['Timestamp'].max()

# Filter the exchange DataFrame to get the row(s) with the latest timestamp
latest_row = exchange_df[exchange_df['Timestamp'] == max_timestamp]

# Extract the USD to EGP exchange rate value from the latest row
USD_to_EGP_value = latest_row['USD_to_EGP']
```

```python
for index, row in order_item_df.iterrows():
   # Assign the USD to EGP exchange rate value to the 'USD_to_EGP' column for each row
    order_item_df.at[index, 'USD_to_EGP'] = USD_to_EGP_value
```

```python
#Calculate list price in Egyptian Pound (EGP) based on the USD to EGP exchange rate
order_item_df['list_price_egp']=order_item_df['list_price']*order_item_df['USD_to_EGP']
```
